```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv("futuristic_city_traffic.csv")
print(df.head())
```

```python
plt.figure(figsize=(8,6))
df.groupby("City")["Speed"].mean().plot(kind="bar", rot=30)
plt.title("Average Speed by City")
plt.ylabel("Speed")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
df.groupby("Vehicle Type")["Traffic Density"].mean().plot(kind="bar", rot=30, color="orange")
plt.title("Average Traffic Density by Vehicle Type")
plt.ylabel("Traffic Density")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Speed"], df["Energy Consumption"], alpha=0.7)
plt.title("Speed vs Energy Consumption")
plt.xlabel("Speed")
plt.ylabel("Energy Consumption")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
df.groupby("Weather")["Traffic Density"].mean().plot(kind="bar", color="green", rot=30)
plt.title("Weather Impact on Traffic Density")
plt.ylabel("Average Traffic Density")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
df["Vehicle Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Distribution of Vehicle Types")
plt.ylabel("")
plt.show()
```
